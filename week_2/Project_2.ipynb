START
SELECT: Option to choose from Quadratic, Cubic, or Quartic equations.

INPUT: Input the values of A, B, C, D, E depending on the choice of equation.

COMPUTE: Solve for the chosen equation.

OUTPUT: Print the roots of the quation.
END

```python
#### import numpy as np

def solve_quadratic():
    print("\nQuadratic Equation: AX^2 + BX + C = 0")
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
    roots = np.roots([a, b, c])
    display_roots(roots)

def solve_cubic():
    print("\n--- Cubic Equation: AX^3 + BX^2 + CX + D = 0 ---")
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
    d = float(input("Enter D: "))
    roots = np.roots([a, b, c, d])
    display_roots(roots)

def solve_quartic():
    print("\nQuartic Equation: AX^4 + BX^3 + CX^2 + DX + E = 0")
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
    d = float(input("Enter D: "))
    e = float(input("Enter E: "))
    roots = np.roots([a, b, c, d, e])
    display_roots(roots)

def display_roots(roots):
    print("\nThe roots are:")
    for i, r in enumerate(roots, 1):
        # Rounding for cleanliness
        clean_root = np.round(r, 4)
        print(f"Root {i}: {clean_root}")

def main():
    while True:
        print("   POLYNOMIAL ROOT FINDER")
        
        print("1. Quadratic Equation")
        print("2. Cubic Equation")
        print("3. Quartic Equation")
        print("4. Exit")
        
        choice = input("\nSelect an operation (1-4): ")

        if choice == '1':
            solve_quadratic()
        elif choice == '2':
            solve_cubic()
        elif choice == '3':
            solve_quartic()
        elif choice == '4':
            print("Program Ended")
            break
        else:
            print("Invalid selection. Please try again.")

if __name__ == "__main__":
    main()
```

```python
print("___Financial Calculator___")

#Simple Interst Calculation
print("Simple Interest Program")
P = float(input("Enter Principal (P): "))
R = float(input("Enter Rate (R): "))
T = float(input("Enter Time (T): "))

A = P * (1 + (R / 100) * T)
print("Simple Interest Amount:", A)


#Compound Interest Calculation
print("Compound Interest Program")
P = float(input("Enter Principal (P): "))
R = float(input("Enter Rate (R): "))
n = float(input("Enter number of times interest applied per period (n): "))
t = float(input("Enter time periods (t): "))

Comp_Interest = P * (1 + R / n) ** (n * t)
print("Compound Interest Amount:", Comp_Interest)


#Annuity Plan Calculation
print("Annuity Plan Program")
PMT = float(input("Enter Payment Amount (PMT): "))
R = float(input("Enter Rate (R): "))
n = float(input("Enter number of times interest applied per period (n): "))
t = float(input("Enter time periods (t): "))

Annuity = PMT * (((1 + R / n) ** (n * t) - 1) / (R / n))
print("Annuity Plan Amount:", Annuity)
```

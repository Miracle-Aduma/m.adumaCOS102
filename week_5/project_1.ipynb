```python
import tkinter as tk
from tkinter import messagebox
import random

class Employee:
    def __init__(self):
        self.employees = [
            "Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali",
            "Andrew Ugwu", "Stella Mankinde", "Jane Akibo", "Ago James",
            "Michell Taiwo", "Abraham Jones", "Nicole Anide",
            "Kosi Korso", "Adele Martins", "Emmanuel Ojo", "Ajayi Fatima"
        ]

        self.tasks = ["Loading", "Transporting", "Reviewing Orders",
                      "Customer Service", "Delivering Items"]

    def check_employee(self, name):
        return name in self.employees

    def take_attendance(self, name):
        return f"Attendance taken for {name}"

    def assign_task(self):
        return random.choice(self.tasks)

    def refuse_access(self):
        return "Access Denied! You are not an employee."


class EmployeeGUI:
    def __init__(self, root):
        self.emp = Employee()
        self.root = root
        self.root.title("Employee System")

        self.label = tk.Label(root, text="Enter Employee Name:")
        self.label.pack()

        self.entry = tk.Entry(root)
        self.entry.pack()

        self.button = tk.Button(root, text="Submit", command=self.process)
        self.button.pack()

        self.result = tk.Label(root, text="")
        self.result.pack()

    def process(self):
        name = self.entry.get()

        if self.emp.check_employee(name):
            attendance = self.emp.take_attendance(name)
            task = self.emp.assign_task()
            self.result.config(text=f"{attendance}\nAssigned Task: {task}")
        else:
            self.result.config(text=self.emp.refuse_access())


# Run GUI
root = tk.Tk()
app = EmployeeGUI(root)
root.mainloop()
```

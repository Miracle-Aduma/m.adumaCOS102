```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("diabetes.csv")
```

```python
# to know what is included and what is not included in your data set
df.info()
```

```python
# to know the data sets with null values  .......... isna means its not
df.isna().sum()
```

```python
# filling data for insulin
# taking the mean of the values in the insulin column and filling the empty spaces with it

#df['Insulin'].fillna(df['insulin'].mean())
```

```python
# To fill empty columns
to_fill = ['BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
for col in to_fill:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.isna().sum()
```

```python
# dropping columns with null values
df.dropna(inplace = True)                            #inplace = True save
```

```python
df.isna().sum()
```

```python
plt.figure(figsize = (10,8))
sns.heatmap(df.corr(), annot = True, cmap = 'coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

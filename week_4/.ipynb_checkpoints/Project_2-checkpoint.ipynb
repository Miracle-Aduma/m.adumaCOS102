```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("titanic.csv")
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
# Creates new cabin
df['have_cabin'] = df['Cabin'].notna().map({True: 'yes', False: 'no'})

df
```

```python
df.isna().sum()
```

```python
# Fills age columns
df['Age'].fillna(df['Age'].mean())
```

```python
# dropping columns with null values
df.dropna(inplace = True)

df.isna().sum()
```

```python
plt.figure()

df['Survived'].value_counts().plot(kind='bar')

plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Number of Passengers')

plt.show()
```

```python
plt.figure()

pd.crosstab(df['Pclass'], df['Survived']).plot(kind='bar')

plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')

plt.show()
```

```python
plt.figure()

pd.crosstab(df['have_cabin'], df['Survived']).plot(kind='bar')

plt.title('Cabin Availability vs Survival')
plt.xlabel('Have Cabin (yes/no)')
plt.ylabel('Number of Passengers')

plt.show()
```

INSIGHTS
The survival count showed that more passengers died than survived (high fatality rate). The second visualization showed that first-class passengers had a significantly higher survival rate compared to third-class passengers.the analysis of cabin availability showed that passengers with cabins were more likely to survive than those without.

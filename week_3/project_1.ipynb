```python
import tkinter as tk
from tkinter import messagebox
import csv


# function to check employee
def check_employee():

    name = name_entry.get()
    department = dept_entry.get()

    employee_found = False
    department_members = []

    # open the csv file
    with open("gig_logistics.csv", "r") as file:
        print("File loaded successfully")
        reader = csv.reader(file)

        next(reader)  # skip header row

        for row in reader:
            emp_name = row[1]
            emp_dept = row[2]
        
            if emp_name.strip().lower() == name.strip().lower() and emp_dept.strip().lower() == department.strip().lower():
                employee_found = True
        
            if emp_dept.strip().lower() == department.strip().lower():
                department_members.append(emp_name)

    if employee_found:
        members = "\n".join(department_members)

        messagebox.showinfo(
            "Employee Found",
            f"Welcome {name}!\n\nMembers in {department} department:\n{members}"
        )
    else:
        messagebox.showinfo(
            "Not Found",
            "Employee does not exist."
        )


# GUI window
window = tk.Tk()
window.title("GIG Logistics Employee Checker")
window.geometry("350x200")

# name label
tk.Label(window, text="Enter Name").pack()

name_entry = tk.Entry(window)
name_entry.pack()

# department label
tk.Label(window, text="Enter Department").pack()

dept_entry = tk.Entry(window)
dept_entry.pack()

# button
tk.Button(window, text="Check Employee", command=check_employee).pack(pady=10)

window.mainloop()
```
